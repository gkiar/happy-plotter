## Here are some examples of changing the color scheme in ggplot2.

The default colors in ggplot sometimes just don't cut it.  Here are some examples of how to pick different ones. 
These are by no means "good" palettes to use, they just give you the exmaple of how to change it to something other than the default.

It's also good to be aware of color blindness when picking a palette.
Here are some links that can help:
[colorbrewer2](http://colorbrewer2.org)
[Adobe](https://color.adobe.com/create/color-wheel/)


```r
suppressMessages(require(ggplot2))
```

## Making some toy data 

```r
dat <- as.data.frame(iris[,-5])
dat$col <- as.factor(kmeans(dat, centers = 3)$cluster)
```

## ggplot default colors

```r
ggplot(dat, aes(x = Sepal.Length,y = Sepal.Width, color = col)) + 
geom_point()
```

## ggplot custom colors

Here's how you add your own color palette to a ggplot.

```r
ColPal1 <- c('#fc8d59','#000000','#91cf60')

ggplot(dat, aes(x = Sepal.Length,y = Sepal.Width, color = col)) + 
    geom_point() + 
    scale_color_manual(values = ColPal2)
```
