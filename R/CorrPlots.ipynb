```r
require(corrplot)
require(foreach)
require(colorRamps)
require(RColorBrewer)
```

```r
M <- 
foreach(x = 1:24, .combine='cbind') %do% {
    A <- rnorm(6)
    B <- rnorm(10, 3, 2)
    C <- runif(4, -5, 5)
    D <- runif(4, -10,10)   
    #row
    c(A,B,C,D)
}
rownames(M) <- LETTERS[1:dim(M)[1]]
colnames(M) <- 1:dim(M)[2]
```

```r
M <- M + t(M)
M <- M/(range(M))
```

```r
cols <- c(rep("blue",6), rep('darkred',10),rep('green',4),rep('purple',4))
tmp <- as.numeric(table(cols))
```

```r
corrplot(M, method='color',tl.col=cols)
corrRect(tmp,col=unique(cols),lwd=4)
```

### Playing with colors

```r
corrplot(M, method='color',tl.col=cols, col = colorRampPalette(brewer.pal(9,"PRGn"))(100))
corrRect(tmp,col=unique(cols),lwd=4)
```

```r
corrplot(M, method='color',tl.col=cols, col = colorRampPalette(brewer.pal(9,"RdYlBu"))(100))
corrRect(tmp,col=unique(cols),lwd=4)
```
