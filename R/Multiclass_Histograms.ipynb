## Multiclass Histograms with prob=TRUE

First we include the libraries needed and create some toy data to play with.  

```r
require(ggplot2)
```

```r
set.seed(1234)
x1 <- rnorm(10000,0,1)
x2 <- rnorm(90000,1,1) 
X <- data.frame(x = c(x1,x2), 
                Class = as.factor(c(rep(1, length(x1)), rep(2, length(x2)))))
```

Now we have toy data which is a mixture model of the form 
$$X \sim P(\cdot; \theta) = 0.1\phi(\cdot; 0, 1) + 0.9\phi(\cdot; 1, 1)$$

```r
n <- dim(X)[1]
bw <- 3.49*sd(X[,'x'])*dim(X)[1]^(-1/3) ## binwidth using Scott's rule.
```

We have computed the bandwidth according to Scott's rule. 
See Scott, D.W. (1992) _Multivariate Density Estimation. Theory, Practice and Visualization._

```r
p1 <- 
  ggplot(data = X, aes(x = x, bw = bw, n = n)) +
    geom_histogram(aes(y = (..count..)/(n*bw), fill=Class), binwidth=bw)  +
    geom_density() 
```

```r
p1
```

Now the bins are colored by the proportion of each class, but the histogram is plotted with prob=TRUE
