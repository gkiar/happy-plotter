## ggplot replacement for heatmap.2

Here's an example.  Colors are purposely bad.

```r
require(ggplot2)
require(gplots)
```

```r
m <- matrix(round(runif(100),2), 10,10)
diag(m) <- 0

m

grid <- expand.grid(y = 1:dim(m)[1],x = 1:dim(m)[2])
grid$v <- as.numeric(m)
```

```r
p <- ggplot(grid, aes(x = x, y = y, fill = v))
p1 <- p + geom_raster() + scale_y_reverse() + 
  geom_text(aes(label = v)) + 
  scale_fill_gradientn(colours = colorpanel(255, "green", "purple")) + 
  guides(fill = FALSE) +
  theme(panel.background = element_blank(), 
        axis.title = element_blank(),
        axis.ticks = element_blank(),
        axis.text = element_blank())
```

```r
p1 + ggtitle("Title Here")
```
