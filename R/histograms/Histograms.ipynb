# Histogram Examples in `ggplot`
Here are some exampeles of different types of ggplot histograms.

```r
## Libraries
suppressMessages(require(ggplot2))
## Data
set.seed(317)
Z <- c(rep(1,1000), rep(5,1000))
X <- data.frame(x = rnorm(length(Z), mean = Z, sd = 1), z = Z)
```

```r
head(X)
```

```r
p <- ggplot(data = X, aes(x = x, fill = as.factor(Z)))
p + geom_histogram(position = 'identity', bins = 50, alpha = 0.5)
p + geom_histogram(alpha = 0.5, bins = 50)
```

Note the difference between the above plots and their generating code. 
The first puts both histograms where you'd expect and this can be seen wtih the appropriate level of transparancy.
The second -- which is default behaviour -- stacks the histograms.
